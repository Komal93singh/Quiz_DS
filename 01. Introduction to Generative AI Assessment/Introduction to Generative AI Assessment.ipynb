### Q11. Which of the following is a key difference between Generative and Discriminative models? 
1. Generative models predict class labels, while discriminative models generate new data 
2. Generative models model the decision boundary, while discriminative models model the data distribution 
3. Generative models learn the joint probability distribution, while discriminative models learn the conditional probability

The correct answer is:

**3. Generative models learn the joint probability distribution, while discriminative models learn the conditional probability**

* **Discriminative models** (e.g., Logistic Regression, SVM) learn **the boundary between classes** and **predict labels** for new data.


### Q12. What is a common challenge in training GANs? 
1. Lack of data 
2. Mode collapse 
3. Overfitting 
4. High computation cost

The correct answer is:

**2. Mode collapse** 

**Mode collapse** occurs when the **generator produces limited varieties of outputs**, ignoring parts of the data distribution. This is a common challenge in GAN training.

Other challenges include **training instability** and **high computation cost**, but mode collapse is the most characteristic issue unique to GANs.

### Q26. In a Hidden Markov Model, what is the purpose of the emission probabilities? 
1. They determine the transition between states 
2. They defines the probability of an observed outcome given a hidden state 
3. They represent the likelihood of the initial state 
4. They measure the accuracy of the model

The correct answer is:

**2. They define the probability of an observed outcome given a hidden state**

In a **Hidden Markov Model (HMM)**:

* **Hidden states** represent the underlying system (not directly observed).
* **Emission probabilities** describe how likely a **particular observation** is given a **specific hidden state**.
* **Transition probabilities** handle the likelihood of moving from one hidden state to another.

### Q27. What is a major challenge when using VAE for image generation? 
1. Mode collapse 
2. Loss of fine details in generated images 
3. High computational complexity 
4. Inability to handle large datasets

The correct answer is:

**2. Loss of fine details in generated images**

In **Variational Autoencoders (VAEs)**:

* This happens because VAEs optimize a **reconstruction loss** combined with a **KL divergence term**, which encourages smooth latent distributions but can **reduce sharpness**.

### Q29. Why is the KL divergence term used in the loss function of Variational Autoencoders (VAE)? 
1. To minimize the difference between the predicted and true labels 
2. To enforce the regularization of the latent space distribution 
3. To ensure the model overfits the training data 
4. To optimize the discriminator network

The correct answer is:

**2. To enforce the regularization of the latent space distribution** 

In **Variational Autoencoders (VAEs)**:

* The **KL divergence term** measures how much the **learned latent distribution** deviates from a **prior distribution** (usually a standard normal).
* This **regularization** ensures that the latent space is **smooth and continuous**, which helps in **generating meaningful and diverse new samples**.


### Q30. What is the primary advantage of using probabilistic models in Generative AI over deterministic models? 
1. Probabilistic models can provide more nuanced predictions by incorporating uncertainty 
2. Probabilistic models are easier to implement 
3. Probabilistic models require less computational power 
4. Probabilistic models always give more accurate predictions

The correct answer is:

**1. Probabilistic models can provide more nuanced predictions by incorporating uncertainty**

* **Probabilistic models** (like HMMs, GMMs, VAEs) capture the **uncertainty and variability in data**, allowing Generative AI to **produce realistic, diverse outputs**.
* In contrast, **deterministic models** produce a single fixed output without modeling uncertainty.
